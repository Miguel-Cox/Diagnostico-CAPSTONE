```python
import pandas as pd
import json
farmers = pd.read_json("prueba.json", lines=True)
farmers.head()
```

```python
def Most_retweeted():
    a = farmers.nlargest(10, 'retweetCount')
    return a
most_retweeted = Most_retweeted()
most_retweeted
```

```python
def Top_users():
    n = 10
    lista = farmers['user'].value_counts()[:n].index.tolist()
    users = pd.DataFrame(lista)
    return users
top_users = Top_users()
top_users
```

```python
def Top_days():
    n = 10
    lista = farmers['date'].dt.date.value_counts()[:n].index.tolist()
    dates = pd.DataFrame(lista)
    return dates
top_days = Top_days()
top_days
```

```python
def Top_hashtags():
    n = 10
    hashtags = []
    for content in farmers['content']:
        if "#" in content:
            for tag in content.split():
                if tag.startswith("#"):
                    hashtags.append(tag)
    dates = pd.DataFrame(hashtags)
    return dates.value_counts()[:n]

top_hashtags = Top_hashtags()
top_hashtags
```
